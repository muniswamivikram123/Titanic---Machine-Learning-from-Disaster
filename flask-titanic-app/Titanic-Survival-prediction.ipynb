```python
import numpy as np
import pandas as pd
```

```python
train_t = pd.read_csv("train.csv")

train_t
```

```python
train_t = train_t[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
train_t
```

```python
train_t = train_t.dropna()
train_t
```

```python
train_t.describe()
```

```python
from tensorflow.keras.utils import to_categorical
pclass = train_t['Pclass'].to_numpy()
as_categorical_pclass = to_categorical(pclass - 1, num_classes=3)
as_categorical_pclass
```

```python
def sex_str_to_binary(s):
    if s=="male":
        return 1
    else:
        return 0
sex = train_t['Sex']

as_binary_sex = sex.apply(sex_str_to_binary).to_numpy()
as_binary_sex[:10]
```

```python
age_np = train_t['Age'].to_numpy()
age_np = (age_np - train_t['Age'].mean()) / train_t['Age'].std()
age_np[:10]
```

```python
sibsp = train_t['SibSp']
sibsp_np = sibsp.to_numpy()
sibsp_np = (sibsp_np - train_t['SibSp'].mean()) / train_t['SibSp'].std()
sibsp_np[:10]
```

```python
parch = train_t['Parch']
parch = parch.to_numpy()
parch = (parch - train_t['Parch'].mean()) / train_t['Parch'].std()
parch[:10]
```

```python
fare = train_t['Fare'].to_numpy()
fare = (fare - train_t['Fare'].mean()) / train_t['Fare'].std()
fare[:10]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
train_Y = train_t['Survived'].to_numpy()
train_Y[:10]
```

```python
train_X = np.concatenate((as_categorical_pclass, 
                as_binary_sex.reshape(-1, 1),
                age_np.reshape(-1, 1),
                sibsp_np.reshape(-1,1),
                parch.reshape(-1,1),
                fare.reshape(-1,1)), axis=1)
train_X
```

```python
model = LogisticRegression(random_state=0).fit(train_X, train_Y)
model.score(train_X, train_Y)
```

```python
import joblib
from sklearn.linear_model import LogisticRegression
from tensorflow.keras.utils import to_categorical
import pandas as pd
import numpy as np

# Load the training data
train_t = pd.read_csv("train.csv")
train_t = train_t[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']].dropna()

def sex_str_to_binary(s):
    return 1 if s == "male" else 0

# One-hot encode Pclass
pclass = to_categorical(train_t['Pclass'] - 1, num_classes=3)
sex = train_t['Sex'].apply(sex_str_to_binary).to_numpy()
age = (train_t['Age'].to_numpy() - train_t['Age'].mean()) / train_t['Age'].std()
sibsp = (train_t['SibSp'].to_numpy() - train_t['SibSp'].mean()) / train_t['SibSp'].std()
parch = (train_t['Parch'].to_numpy() - train_t['Parch'].mean()) / train_t['Parch'].std()
fare = (train_t['Fare'].to_numpy() - train_t['Fare'].mean()) / train_t['Fare'].std()

# Combine features
train_X = np.hstack((pclass, sex.reshape(-1, 1), age.reshape(-1, 1), sibsp.reshape(-1, 1), parch.reshape(-1, 1), fare.reshape(-1, 1)))
train_Y = train_t['Survived'].to_numpy()

# Train the model
model = LogisticRegression(random_state=0).fit(train_X, train_Y)

# Save the model
joblib.dump(model, 'titanic_model.joblib')
print("Model saved successfully!")
```
